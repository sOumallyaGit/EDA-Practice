```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Beacuse File is present in excel Formate
df = pd.read_excel('flight_price.xlsx')
```

```python
df.head()
```

```python
## Getting Basic Info about data
df.info()
```

```python
df.describe()
```

```python
## Feature Engineering Handling Data Column
df["date"]=df["Date_of_Journey"].str.split("/").str[0]
df["month"]=df["Date_of_Journey"].str.split("/").str[1]
df["year"]=df["Date_of_Journey"].str.split("/").str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
## So the new columns are still in Object formate so we need to convert them into Int type
df["date"]=df["date"].astype(int)
df["month"]=df["month"].astype(int)
df["year"]=df["year"].astype(int)
```

```python
df.info()
```

```python
## Now we are dropping the date of journey Column
df.drop("Date_of_Journey",axis=1,inplace=True)
```

```python
df.info()
```

```python
## Handling the arrival time column
## getting only the time info in dataset
df["Arrival_Time"]=df["Arrival_Time"].str.split(" ").str[0]
```

```python
## Separating Hour and mintues and also converting them to integer
df["Arrival_hour"]=df["Arrival_Time"].str.split(":").str[0].astype(int)
df["Arrival_min"]=df["Arrival_Time"].str.split(":").str[1].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
## dropping the original Arrival Time Column
df.drop("Arrival_Time",axis=1,inplace=True)
```

```python
df.head()
```

```python
## Handling Departure Time Column spliting and converting to int
df["Dep_hour"]=df["Dep_Time"].str.split(":").str[0].astype(int)
df["Dep_min"]=df["Dep_Time"].str.split(":").str[1].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
## Dropping Dep_Time Column
df.drop("Dep_Time",axis=1,inplace=True)
```

```python
df.head()
```

```python
## Working On Categorical Values
## Total_Stops

df["Total_Stops"].unique()
```

```python
df["Total_Stops"].value_counts()
```

```python
## Finding The Number of missing Values
df["Total_Stops"].isnull().sum()
```

```python
## Finding The actual Column where null Value is present
df[df["Total_Stops"].isnull()]
```

```python
## Handling The missing Values with Mode Value
df["Total_Stops"].mode()
```

```python
## So as the value of mode is 1 Stop so we are going to encode nan with 1 stop value
df["Total_Stops"]=df["Total_Stops"].map({"non-stop":0,"1 stop":1,"2 stops":2,"3 stops":3,"4 stops":4,np.nan:1})
```

```python
df[df["Total_Stops"].isnull()]
```

```python
df.head()
```

```python
df.info()
```

```python
## Dropping The route Column
df.drop("Route",axis=1,inplace=True)
```

```python
df.head()
```

```python
## Handling The Duration Colum First Hour
# Extract hours, safely convert to integers
df["duration_hour"] = df["Duration"].apply(lambda x: int(x.split("h")[0]) if "h" in x else 0)
```

```python
## Handling The Duration Column Minutes
df["duration_min"] = df["Duration"].apply(lambda x: x.split(" ")[1].split("m")[0] if len(x.split(" ")) > 1 else "0")
```

```python
df.head()
```

```python
df.info()
```

```python
## Converting Duration hour and mintues to integer type
df["duration_hour"]=df["duration_hour"].astype(int)
df["duration_min"]=df["duration_min"].astype(int)
```

```python
df.info()
```

```python
## Dropping The duration Column
df.drop("Duration",axis=1,inplace=True)
```

```python
df.head()
```

```python
## Handling Categorical Features
```

```python
df["Airline"].unique()
```

```python
df["Airline"].value_counts()
```

```python
## ONE HOT ENCODING ALL Categorical FEATURE
from sklearn.preprocessing import OneHotEncoder
OHE = OneHotEncoder()
enoded_df=pd.DataFrame(OHE.fit_transform(df[["Airline","Source","Destination"]]).toarray(),columns=OHE.get_feature_names_out())
```

```python
## Concatinating two df
new_df=pd.concat([df,enoded_df],axis=1)
```

```python
new_df.head()
```

```python
## Dropping The Categorical Features that were encoded
new_df.drop(["Airline","Source","Destination"],axis=1,inplace=True)
```

```python
new_df.head()
```

```python
new_df.info()
```

```python
## Dropping Additonal INfo Category
new_df["Additional_Info"].unique()
```

```python
new_df.drop("Additional_Info",axis=1,inplace=True)
```
